## Cleaning Method

```python
print(df.isnull().sum())
```

```python
df.drop(columns=['fulfilled-by', 'promotion-ids', 'Unnamed: 22'], errors='ignore', inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df['Courier Status'].fillna("Not Provided", inplace=True)
```

```python
df['currency'].fillna("USD",inplace=True)
```

```python
df['Amount'].fillna(df['Amount'].median(), inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['ship-city','ship-state','ship-postal-code','ship-country'], inplace=True)
```

```python
df.isnull().sum()
```

## Convert Dates

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

## Step 6: Basic Analysis & Visualization

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

### Sales by Category

```python
plt.figure(figsize=(10,6))
df.groupby('Category')['Amount'].sum().sort_values(ascending=False).plot(kind='bar', color='orange')
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.show()
```

### Sales by Category

```python
plt.figure(figsize=(12,6))
df.groupby('Month')['Amount'].sum().plot(kind='line', marker='o',color='orange')
plt.title("Monthly Sales Trend")
plt.ylabel("Total Sales")
plt.xlabel("Month")
plt.show()
```

### Top 10 Products by sales

```python
plt.figure(figsize=(12,6))
df.groupby('SKU')['Amount'].sum().sort_values(ascending=False).head(10).plot(kind='bar', color='orange')
plt.title("Top 10 Products by Sales")
plt.ylabel("Total Sales")
plt.show()
```
