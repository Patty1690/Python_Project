```python
# drop duplicates
titanic.drop_duplicates(inplace=True)
titanic.reset_index(inplace= True, drop= True)
titanic
```

```python
# Checking null values
titanic.isnull()

null_counts = titanic.isnull().sum()
if null_counts.sum()>0:
    print(null_counts[null_counts>0])
```

Data Cleaning and missing values

```python
# If the number of entries is N , 5-10% missing values-mwe can eliminate the rows
titanic = titanic[titanic['Embarked'].notna()]

# If the number of entries is N , 30-50% missing values - we can replace the rows with mean, median, mode

titanic['Age'].fillna(titanic['Age'].median(),inplace= True)

# If the number of entries is N , 70% missing values - we sholud eliminate the column
titanic= titanic.drop('Cabin', axis=1)

titanic
```

```python
titanic['Age']=titanic['Age'].astype('int64')
```

```python
# Check outliers
import seaborn as sns
sns.boxplot(titanic['Fare'])
```

```python
titanic= titanic[titanic.Fare<500]
```

```python
import numpy as np
z_scores = np.abs((titanic[['Age','SibSp']]-titanic[['Age','SibSp']].mean())/titanic[['Age','SibSp']].std())

#ZScore
outliers_threshold = 3
outliers = titanic[(z_scores['Age']>outliers_threshold) |(z_scores['SibSp']>outliers_threshold)]
print(outliers)
```

```python
# Winsorise the outliers by replacing them with the 95th percentile value
age_threshold = titanic['Age'].quantile(0.95)
SibSp_threshold= titanic['SibSp'].quantile(0.95)

titanic['Age'] = np.where(titanic['Age']> age_threshold, titanic['Age'])
titanic['SibSp'] = np.where(titanic['SibSp']> SibSp_threshold, titanic['SibSp'])
```

```python
# drop columns
titanic.drop(['PassengerId', 'Name','Ticket'], axis=1, inplace=True)
titanic
```

```python
# Group by

survived = titanic.groupby('Survived').mean()
survived
```

```python
# Filtering
female_passenger = titanic[titanic.Sex=='female']
female_passenger
```

```python
# Visualization

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.histplot(data=titanic,x ='Age',hue='Survived', bins = 30, kde=True)
plt.title("Distribution of passenger by Age and survival")
plt.show()
```
